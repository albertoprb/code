```python
import pandas as pd
# read the udemy json at ../data/udemy_development_courses.json into a data frame
df = pd.read_json('../data/udemy_development_courses.json')
df.head(5)
```

### Setting the primary key clearly

```python
# Rename ID column
df = df.rename(columns={'id': 'udemy_id'})
df.info()
```

### Dropping unnecessary columns

```python
# Drop columns that have no meaningful distinctive values
df = df.drop(columns=['is_paid'])
df = df.drop(columns=['has_certificate'])
df.info()
```

### Flattening pricing column

```python
# Extract price from JSON
df_price_detail = pd.json_normalize(df['price_detail'])
df['amount'] = df_price_detail['amount']
df['currency'] = df_price_detail['currency']
df = df.drop(columns=['price_detail'])
df.head(3)
```

```python
# Adjust course duration columns
df['content_info'] = df['content_info'].str.replace("total hours", "", regex=False)
df = df.rename(columns={'content_info': 'content_length_hours'})
df = df.rename(columns={'estimated_content_length': 'content_length_minutes'})
df.head(3)
```

### Flattening instructors lists and keeping only the name

```python
# Determine the maximum number of items in the 'Instructor' column
max_items = df['visible_instructors'].apply(len).max()

# Create new column names for the Instructors
new_columns = [f'Instructor{i+1}' for i in range(max_items)]


# Splitting the list in 'Items' column into separate columns
df[new_columns] = pd.DataFrame(df['visible_instructors'].tolist(), index=df.index)
for column in new_columns:
    df[column] = df[column].apply(lambda x: x['title'] if x is not None and 'title' in x else None)


# Drop the original 'visible_instructors' column
df = df.drop('visible_instructors', axis=1)

df
```

### Flatten Locale

```python
# Using only the title of locale as reference
df_locale = pd.json_normalize(df['locale'])['title']
# Assigning the locale the processed value
df['locale'] = df_locale
df
```

### Transform categories

```python
# Using title for the categories
df_primary_category = pd.json_normalize(df['primary_category'])['title']
df_primary_subcategory = pd.json_normalize(df['primary_subcategory'])['title']

# Assigning the categories to the processed values
df['primary_category'] = df_primary_category
df['primary_subcategory'] = df_primary_subcategory
df = df.rename(columns={'primary_category': 'category'})
df = df.rename(columns={'primary_subcategory': 'subcategory'})
df
```

### Separate content fields

I'll separate the text, url, image fields for:

- NLP analysis later based on use cases after the EDA 
- Serving as a 'database' to show course details in charts

They are not necessary for stastical analysis as it's not categorical nor numerical information.


```python
df['requirements_data'].value_counts()
df_text = df[['udemy_id', 
              'title', 
              'url', 
              'description',
              'headline',
              'image',
              'requirements_data', 
              'what_you_will_learn_data', 
              'target_audiences', 
              'objectives']]

# Dropping all columns from the the dataframe except the udemy_id   
df= df.drop(columns=['title', 
              'url', 
              'description',
              'headline',
              'image',
              'requirements_data', 
              'what_you_will_learn_data', 
              'target_audiences', 
              'objectives'])

df
```

```python
df_text
```

### Transform labels in categorical information

```python
def keep_only_label_title(labels):
    label_titles = []
    for label in labels:
        label_titles.append(label['title'])
    return label_titles

df['labels'] = df['labels'].apply(keep_only_label_title)
df
```

### Converting description from HTML to text

```python
from bs4 import BeautifulSoup

def html_to_string(html):
    soup = BeautifulSoup(html, 'html.parser')
    return soup.get_text()

df_text['description'].apply(html_to_string)

df_text
```

### Export results to CSV
